# Clustering

Identifying the similar group of data in a data set is called clustering.Notion of clustering cannot be precisely defined.

Clustering is classified into 2 types:hard clustering and soft clustering.

Following are the types of clustering algorithms:connectivity models(hierarchical clustering algorithm),centroid models(K-means),distribution models() and density models( DBSCAN and OPTICS).


K-means Algorithm:

K-means belongs to centroid model

K means is an iterative clustering algorithm that aims to find local maxima in each iteration. This algorithm works in these 5 steps :

Firstly we need to specify desired number of clusters.Let it be k.Let us choose k=2 for these 5 data points in 2-D space.
    
Randomly assign each data point to a cluster.Few points in 1 cluster and other points in second cluster.

Calculate centroids for each cluster.

Re-assign each point to the closest cluster centroid.

This clustering is an iterative process.Group the nearest points make them as cluster and calculate centroid till there will be no switching of the data points between the clusters.

```python
import pandas as p

import pylab as pl#for plotting graphs

from sklearn.cluster import KMeans #for clustering algorithm

from sklearn.decomposition import PCA#principle component analysis
```

```python
variables = p.read_csv('sample_stocks.csv')

Y = variables[['returns']]

X = variables[['dividendyield']]
```

```python
Nc = range(1, 10)

kmeans = [KMeans(n_clusters=i) for i in Nc]

kmeans

score = [kmeans[i].fit(Y).score(Y) for i in range(len(kmeans))]

score

pl.plot(Nc,score)

pl.xlabel('Number of Clusters')

pl.ylabel('Score')

pl.title('Elbow Curve')

pl.show()
```

```python
pca = PCA(n_components=1).fit(Y)

pca_d = pca.transform(Y)

pca_c = pca.transform(X)
```

pca is used in order to interpret data easily(to make the overly dispersed data to linear form).

```python
kmeans=KMeans(n_clusters=3)

kmeansoutput=kmeans.fit(Y)

kmeansoutput

pl.figure('3 Cluster K-Means')

pl.scatter(pca_c[:, 0], pca_d[:, 0], c=kmeansoutput.labels_)

pl.xlabel('Dividend Yield')

pl.ylabel('Returns')

pl.title('3 Cluster K-Means')

pl.show()
```

Stocks having higher dividend yeilds also have high returns
