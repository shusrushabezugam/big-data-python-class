# Dimensionality Reduction

Dimensionality of data according to statistics refers to how many attributes a dataset has.

There are  factors based on which classification.They are variables and features.
Dimensionality reduction is the process of reducing the number of random variables under consideration, by obtaining a set of principal variables.

Two components of dimensionality reduction:
    
Feature Selection:Smaller subset from orginal data which is helpfulin modeling the problem
    
Feature Extraction:It reduces high dimensional data to low dimensional data.

Methods of Dimensionality Reduction

The various methods used for dimensionality reduction include:

Principal Component Analysis (PCA)
Linear Discriminant Analysis (LDA)
Generalized Discriminant Analysis (GDA)

# Initial Data setup

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

# Reduction

each digit has 8x8 pixcels implies each digit has 64 dimensions.Now we shall apply dimensionality reduction on this and reduce dimensions from 64 to 2 using PCA

```python
pca = PCA(2)  # project from 64 to 2 dimensions
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
Our reduced dataset is projected with 2 dimensions.
```

# Graph

We can now plot the first two principal components of each point to learn about the data:

```python

plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('spectral', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
plt.show()
```

Above graph represents all digits with 64 dimensions in 2 dimensional graph.
Recall what these components mean: the full data is a 64-dimensional point cloud, and these points are the projection of each data point along the directions with the largest variance. Essentially, we have found the optimal stretch and rotation in 64-dimensional space that allows us to see the layout of the digits in two dimensions, and have done this in an unsupervised manner—that is, without reference to the labels.
